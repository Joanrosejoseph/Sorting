# **Sorting**

## **<u>Insertion Sort**

```python
import pandas as pd
```

```python
url1 = "https://raw.githubusercontent.com/Joanrosejoseph/datasets/master/kochi_metro_frequent_travellers.csv"
```

```python
df = pd.read_csv(url1)
```

```python
df.head()
```

```python
def insertion_sort(arr, key):
    for i in range(1, len(arr)):
        current = arr[i]
        j = i - 1
        while j >= 0 and arr[j][key] > current[key]:
            arr[j + 1] = arr[j]
            j -= 1
        arr[j + 1] = current
    return arr
```

```python
records = df.to_dict('records')
```

```python
sorted_records = insertion_sort(records, key='Trips_Per_Month')
```

```python
sorted_df = pd.DataFrame(sorted_records)
print('\nSorted dataset (by Trips_Per_Month):')
sorted_df
```

## **<u>Merge Sort**

```python
url2 = "https://raw.githubusercontent.com/Joanrosejoseph/datasets/master/metro_card_users_frequent_travellers_zones.csv"
```

```python
df = pd.read_csv(url2)
```

```python
print('Original dataset (first 5 rows):')
df.head()
```

```python
def merge_sort(arr, key):
    if len(arr) > 1:
        mid = len(arr) // 2
        left_half = arr[:mid]
        right_half = arr[mid:]

        merge_sort(left_half, key)
        merge_sort(right_half, key)

        i = j = k = 0

        while i < len(left_half) and j < len(right_half):
            if left_half[i][key] <= right_half[j][key]:
                arr[k] = left_half[i]
                i += 1
            else:
                arr[k] = right_half[j]
                j += 1
            k += 1

        while i < len(left_half):
            arr[k] = left_half[i]
            i += 1
            k += 1

        while j < len(right_half):
            arr[k] = right_half[j]
            j += 1
            k += 1
```

```python
records = df.to_dict('records')

merge_sort(records, key='Trips_Per_Week')
```

```python
sorted_df = pd.DataFrame(records)
print('\nSorted dataset (by Trips_Per_Week):')
sorted_df
```
